## loc / iloc 를 활용한 열 추출 / 행 추출
- 두가지의 차이 기억하기
- 못하겠으면 loc라도

```python
df_KTX.loc['2013':'2016']
```

```python
#df_KTX.loc[행,열]  < 기준임

df_KTX.loc['2013':'2016', '경부선 KTX']
```

```python
#df_KTX.loc[행조건식, [컬럼명]] < 기준임

df_KTX.loc['2013':'2016', ['호남선 KTX']]
```

```python
iris = sns.load_dataset('iris')

# sepal_length의 길이가 7cm 이상인 것만 조회
# iris.loc[행조건,:] :<< 는 모든 데이터를 가져오는것

iris.loc[iris['sepal_length'] >= 7, :]
iris
```

```python
#species 가 virginica 인것만 추출

iris.loc[iris['species'] == 'virginica', :].head(1)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회

var = 'total_bill'
tips.loc[tips[var] <= tips[var].median(), :]

# tip의 평균보다 크면서, time이 dinner인 데이터만 조회
```

```python
# day가 sun인 데이터만 조회

tips.loc[tips['day'] == 'Sun', :]
```

```python
# And = &
# Or  = | (슆트 \)

tips.loc[(tips['tip'] >= tips['tip'].mean()) & (tips['time'] =='Dinner')] # & (tips[size] == 3) 하면 사이즈가 3인 조건까지 거는거임]  # 다중조건 예시
```
