## GroupBy 연산
- SQL : Group By와 Logic은 동일함
- Groupby 기준컬럼 : 패턴이 있는 문자열
- Groupby 대상컬럼 : 수치 데이터
- 필요한 함수 : 집계함수 (합, 평균, 표준편차, 최대, 최소등)

```python
#            기준        대상     집계
df.groupby('Location')['매출액'].agg(['sum', 'mean', 'std'])
```

```python
df.groupby('Location')['매출액'].agg(['sum', 'mean', 'std']).reset_index()
```

```python
import numpy as np

speeds = pd.DataFrame(
    [
        ("bird", "Falconiformes", 389.0),
        ("bird", "Psittaciformes", 24.0),
        ("mammal", "Carnivora", 80.2),
        ("mammal", "Primates", np.nan),
        ("mammal", "Carnivora", 58),
    ],
    index=["falcon", "parrot", "lion", "monkey", "leopard"],  # 데이터를 저장하는 리스트의 인덱스를 지정
    columns=("class", "order", "max_speed"),
)

speeds
```

```python
grouped = speeds.groupby("class")  # speeds라는 데이터프레임을 class 열을 기준으로 그룹화하는 것입니다.

grouped = speeds.groupby(["class", "order"])  # speeds 데이터프레임을 class 열과 order 열을 기준으로 함께 그룹화
```

```python
import pandas as pd

df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)


df
```

```python
df.groupby("A")

grouped = df.groupby(["A", "B"])
```

```python
df2 = df.set_index(["A", "B"])

grouped = df2.groupby(level=df2.index.names.difference(["B"]))

grouped.sum()
```

```python
def get_letter_type(letter):
    if letter.lower() in 'aeiou':
        return 'vowel'
    else:
        return 'consonant'

grouped = df.T.groupby(get_letter_type)
```

```python
lst = [1, 2, 3, 1, 2, 3]

s = pd.Series([1, 2, 3, 10, 20, 30], lst)

grouped = s.groupby(level=0)

grouped.first()
```
