```python
df_A_B = pd.DataFrame({'판매월': ['1월', '2월', '3월', '4월'],
                       '제품A': [100, 150, 200, 130],
                       '제품B': [90, 110, 140, 170]})

df_C_D = pd.DataFrame({'판매월': ['1월', '2월', '3월', '4월'],
                       '제품C': [112, 141, 203, 134],
                       '제품D': [90, 110, 140, 170]})

df_C_D
```

```python
df_left = pd.DataFrame({'key':['A','B','C'], 'left': [1, 2, 3]})
df_right = pd.DataFrame({'key':['A','B','D'], 'right': [4, 5, 6]})
```

```python
df_left.merge(df_right, how='left', on = 'key')
```

```python
df_left.merge(df_right, how='right', on = 'key')
```

```python
df_left.merge(df_right, how='outer', on = 'key') # FULL JOIN
```

```python
df_left.merge(df_right, how='inner', on = 'key') # 교집합
```

```python
import pandas as pd

air_quality_no2 = pd.read_csv("air_quality_no2_long.csv",
                              parse_dates=True)

air_quality_no2 = air_quality_no2[["date.utc", "location",
                                   "parameter", "value"]]

air_quality_no2.head()            # 데이터 불러와서 특정 열 선택해서 데이터 뽑기
```

```python
air_quality_pm25 = pd.read_csv("air_quality_pm25_long.csv",
                               parse_dates=True)

air_quality_pm25 = air_quality_pm25[["date.utc", "location",
                                     "parameter", "value"]]

air_quality_pm25.head()         # 데이터 불러와서 특정 열 선택해서 데이터 뽑기
```

```python
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)

air_quality.head()  # air_quality_pm25,air_quality_no2 수직으로 합치기
```

```python
print('Shape of the ''air_quality_pm25' table: ', air_quality_pm25.shape)
print('Shape of the ``air_quality_no2`` table: ', air_quality_no2.shape)
print('Shape of the resulting ``air_quality`` table: ', air_quality.shape)

 # 뭐임이거 왜안돼

#print('Shape of the ``air_quality_pm25`` table:', air_quality_pm25.shape)
#print('Shape of the ``air_quality_no2`` table:', air_quality_no2.shape)
#print('Shape of the resulting ``air_quality`` table:', air_quality.shape)
```

```python
air_quality = air_quality.sort_values("date.utc")

air_quality.head()  # 데이터 정렬
```

```python
air_quality = pd.read_csv("air_quality_no2.csv", index_col=0, parse_dates=True)

air_quality.head()       # 파일 불러와서 데이터프레임에 저장
```

```python
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882

air_quality.head()        # 새로운 열 추가하는거 station_london" 열의 값을 * 1.882 london_mg_per_cubic 열에 넣기
```

```python
air_quality["ratio_paris_antwerp"] = (
    air_quality["station_paris"] / air_quality["station_antwerp"]
)


air_quality.head()               # station_paris의 값을 station_antwerp로 나눈값을 구하는것
```
